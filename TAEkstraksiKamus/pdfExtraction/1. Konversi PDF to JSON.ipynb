```python
import fitz
import json
import os
import pandas as pd
```

```python
class BytesEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, bytes):
            return obj.decode('utf-16')
        return json.JSONEncoder.default(self, obj)
```

```python
from pathlib import Path
directory = "../[Full] Daftar Kamus Ekstraksi"
directory_hasil = "../[Full] Bentuk JSON"
pdfgagal = []

os.makedirs(directory_hasil, exist_ok=True)

for filename in os.listdir(directory):
    print("=== Processing File " + filename + " ===")
    try:
        filepath = os.path.join(directory, filename)
        document = fitz.open(filepath)
        
        new_filename = os.path.splitext(filename)[0]
        output_path = os.path.join(directory_hasil, new_filename + "-hasil.json")
        
        with open(output_path, "w", encoding="utf-8") as t:
            counter = 0
            result = {}
            for page in document:
                data = {}
                value = page.get_text('dict')
                for block in value.get('blocks', []):
                    if 'image' in block:
                        del block['image']
                
                data["page"] = counter
                data["text"] = value
                result[str(counter)] = data
                counter += 1
            
            json.dump(result, t, cls=BytesEncoder)
            
        document.close()
    except Exception as e:
        print("=== File Gagal " + filename + " ===")
        print(e)
        pdfgagal.append(filename)
```

```python
for i in pdfgagal:
    print(i)
```
