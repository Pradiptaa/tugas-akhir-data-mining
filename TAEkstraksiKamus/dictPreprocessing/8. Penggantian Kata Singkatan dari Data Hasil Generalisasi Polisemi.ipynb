```python
import pandas as pd
import numpy as np
from ast import literal_eval
import os
```

```python
def to_entry(ls):
    ret = ''
    for isi in ls:
        ret += ' '
        ret += str(isi[0])
    return ret[1:]

def rapi(s):
    ret = ''
    for ch in str(s):
        if ch == '-' or ch.isalpha() or ch == ' ':
            ret += ch
    return ret.strip().lower()
```

```python
def replace_word(df, nomor, singkatan_df):
    replace_df = singkatan_df[singkatan_df['nomor'] == int(nomor)]
    dct = {}
    for i, row in replace_df.iterrows():
        dct[row['singkatan']] = row['kepanjangan']
        
    for i, entry in df.iterrows():
        list_entry = eval(str(entry['entry_font_size_pos']))
        new_list_entry = []
        for idx, ent in enumerate(list_entry):
            if rapi(ent[0]) in dct:
                list_entry[idx][0] = str(list_entry[idx][0]).replace(rapi(ent[0]), str(dct[rapi(ent[0])]))
            if len(str(list_entry[idx][0])) > 0:
                new_list_entry.append(list_entry[idx])
                
        df.loc[i, 'entry_font_size_pos'] = str(new_list_entry)
        df.loc[i, 'Entri'] = to_entry(new_list_entry)
        
    return df
```

```python
base_dir = "../Ekstraksi"
singkatan_path = os.path.join(base_dir, "X. Data Pendukung", "Singkatan Kamus.csv")
singkatan = pd.read_csv(singkatan_path)

input_dir = os.path.join(base_dir, "2. Preprocessing - Generalisasi Polisemi")
output_dir = os.path.join(base_dir, "3. Preprocessing - Replace Kata Singkatan")

os.makedirs(output_dir, exist_ok=True)

cnt = 0
for filename in os.listdir(input_dir):
    if filename.endswith('.csv') and filename[0].isnumeric():
        nomor = filename.split('_')[0]
        f = os.path.join(input_dir, filename)
        df = pd.read_csv(f)
        
        df_replace = replace_word(df, nomor, singkatan)
        
        output_filename = f"{nomor}_Replace.csv"
        df_replace.to_csv(os.path.join(output_dir, output_filename), index=False)
        
        cnt += 1
        print(f"Done Kamus {nomor} Total {cnt} Kamus")
```
