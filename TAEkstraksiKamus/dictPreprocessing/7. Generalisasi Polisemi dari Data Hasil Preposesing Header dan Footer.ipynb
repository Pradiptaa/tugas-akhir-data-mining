```python
import pandas as pd
import numpy as np
from ast import literal_eval
import os
```

```python
def extract_names(ls):
    ret = []
    cur_num = 1
    tmp = []
    for isi in ls:
        try:
            isi[0].isnumeric()
            if isi[0].isnumeric() and int(isi[0]) == cur_num and tmp != []:
                cur_num += 1
                ret.append(tmp)
                tmp = []
            else:
                tmp.append(isi)
        except Exception:
            tmp.append(isi)
    if tmp != []:
        ret.append(tmp)
    return ret
```

```python
def to_entry(ls):
    ret = ''
    for isi in ls:
        ret += ' '
        ret += str(isi[0])
    return ret[1:]
```

```python
import re
from ast import literal_eval

def safe_parse_numpy(val):
    if type(val) != str:
        return val
    
    clean_str = re.sub(r"np\.\w+\(([^)]+)\)", r"\1", val)
    
    try:
        return literal_eval(clean_str)
    except Exception:
        return val
```

```python
def parse_polisemi(df):
    entri = []
    entry_font_size_pos = []
    posisi = []
    page = []
    is_padanan_lema = []
    for i, entry in df.iterrows():
        ls = extract_names(safe_parse_numpy(str(entry['entry_font_size_pos'])))
        if len(ls) > 2:
            for isi in ls[1:]:
                entry_font_size_pos.append(str(ls[0] + isi))
                posisi.append(entry['posisi_entry'])
                page.append(entry['page'])
                is_padanan_lema.append(entry['is_padanan_lema'])
        else:
            entry_font_size_pos.append(str(entry['entry_font_size_pos']))
            posisi.append(entry['posisi_entry'])
            page.append(entry['page'])
            is_padanan_lema.append(entry['is_padanan_lema'])
            
    for isi in entry_font_size_pos:
        entri.append(to_entry(safe_parse_numpy(isi)))
        
    ret = pd.DataFrame()
    ret['Entri'] = entri
    ret['entry_font_size_pos'] = entry_font_size_pos
    ret['posisi_entry'] = posisi
    ret['page'] = page
    ret['is_padanan_lema'] = is_padanan_lema
    return ret
```

```python
base_dir = "../Ekstraksi"
input_dir = os.path.join(base_dir, "1. Preprocessing - Header _ Footer")
output_dir = os.path.join(base_dir, "2. Preprocessing - Generalisasi Polisemi")

os.makedirs(output_dir, exist_ok=True)

cnt = 0
for filename in os.listdir(input_dir):
    if filename.endswith('.csv') and filename[0].isnumeric():
        nomor = filename.split('.')[0]
        f = os.path.join(input_dir, filename)
        df = pd.read_csv(f)
        
        df_polisemi = parse_polisemi(df)
        
        output_filename = f"{nomor}_Parse.csv"
        df_polisemi.to_csv(os.path.join(output_dir, output_filename), index=False)
        
        cnt += 1
        print(f'Done Kamus {nomor} (Total {cnt} Kamus)')
```
