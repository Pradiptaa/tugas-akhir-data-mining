```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
```

```python
def getBilingualLexicon(df):
    return df[df['is_anomali'] == 0][['kata_asal', 'kata_tujuan', 'makna']]
```

```python
def getMorfology(df):
    kata = []
    form = []
    tag = []

    main_entry = ''
    for i, row in df.iterrows():
        exist = 0
        if row['main_entry'] == 1:
            main_entry = row['kata_asal']
            if row['is_anomali']:
                main_entry = '-'
        elif row['kata_turunan'] == 1 and main_entry != '-':
            kata.append(main_entry)
            form.append(row['kata_asal'])
            tag.append(row['POS'])

    ret = pd.DataFrame()
    ret['kata'] = kata
    ret['form'] = form
    ret['tag'] = tag
    return ret
```

```python
import os
from pathlib import Path
directory = "../Ekstraksi/"
path_billex = directory + '9. Bilingual Lexicon/'
path_morfology = directory + '10. Morphology/'
cnt = 0
daftar_kamus = [18,68,71,54,89]
for filename in os.listdir(directory + '8. Cleaning Informasi Sumber Daya NLP/'):
    if(filename[-3:] == 'csv'):
        f = os.path.join(directory + '8. Cleaning Informasi Sumber Daya NLP/', filename)
        nomor = filename.split('_')[0]
        # if int(nomor) not in daftar_kamus:
            # continue
        df = pd.read_csv(f)
        df_clean = df
        df_bilingual_lexicon = getBilingualLexicon(df_clean)
        df_bilingual_lexicon.to_csv(path_billex + nomor + "_Billex.csv", index = False)
        df_morfology = getMorfology(df_clean)
        df_morfology.to_csv(path_morfology + nomor + "_Morphology.csv", index = False)
        cnt += 1
        # print('Done Kamus', nomor, '(Total', cnt, 'Kamus)')
        print(nomor, len(df_morfology), len(df_bilingual_lexicon), sep = '\t')
```
